```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load the dataset
file_path = '/content/library-facility.csv'  # Replace with the correct path
library_data = pd.read_csv(file_path)

# Selecting only the numerical columns for the analysis
numerical_data = library_data.select_dtypes(include=['int64', 'float64'])

# Standardizing the data
scaler = StandardScaler()
normalized_data = scaler.fit_transform(numerical_data)
```

```python
from sklearn.decomposition import PCA

# PCA
pca = PCA(n_components=2)
pca_results = pca.fit_transform(normalized_data)
```

```python
import matplotlib.pyplot as plt

# PCA
pca = PCA(n_components=2)
pca_results = pca.fit_transform(normalized_data)

# Visualization
plt.figure(figsize=(8, 6))
plt.scatter(pca_results[:, 0], pca_results[:, 1])
plt.title('PCA Results')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

```python
from sklearn.decomposition import TruncatedSVD

# SVD
svd = TruncatedSVD(n_components=2)
svd_results = svd.fit_transform(normalized_data)
```
